```python
import pandas as pd
```

```python
fnamea='../out_n.txt'
fnameb='../out_t.txt'
```

```python
dfn=pd.read_csv('../out_n.txt')
dft=pd.read_csv('../out_t.txt')
```

```python
dfn=dfn[col]
dft=dft[col]
```
