```python
human_prefix = 'AA_Human_'
swine_prefix = 'AA_Swine_'

hpat = r'AA_(Human)_([0-9]{4})([0-9]{4})'
spat = r'AA_(Swine)_([0-9]{4})([0-9]{4})'

def year_to_swine(year):
    return swine_prefix + str(year) + str(year+1)

def year_to_human(year):
    return human_prefix + str(year) + str(year+1)

def gen_path(year, human=False, file_path = '../../data/database_may30_2017/'):
    if human:
        return os.path.abspath(os.path.join(file_path, year_to_human(year)))
    else:
        return os.path.abspath(os.path.join(file_path, year_to_swine(year)))

RESPONSE = 'SPECIES'
SP1 = 'Human'
SP2 = 'Swine'
NUMTREE = 300
PLOT = False
```

```python
def featimp(year):
    FILE = gen_path(year, human=True)
    FILE2 = gen_path(year)

    datatrain = ml.setdataframe(FILE, FILE2, disable_na = True)

    RF,PrRF,ACCRF,CFRF,PrxRF,ACCxRF,CFxRF,RFimp,EFI = ml.randomForestX(RESPONSE__ = RESPONSE, datatrain__ = datatrain, cores=16)
    
    return (year, EFI, RF,PrRF,ACCRF,CFRF,PrxRF,ACCxRF,CFxRF,RFimp)
```
