```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 36}

mpl.rc('font', **font)
```

```python
df=pd.read_csv('out_1.txt',delimiter=",",index_col=None,
                   engine='python')
```

```python
df.to_csv('temp.csv',index=False)
```

```python
dfc=df.CELL.value_counts()
dfc.plot(kind='bar')
```

```python
sns.set_style('dark')
sns.set(rc={'axes.facecolor':'.4', 'figure.facecolor':'.3'})

dfg=df.MYC.value_counts()
ax=dfg.plot(kind='bar',color='g',legend=False)
ax.set_title('MYC',color='w',fontsize=36)
[i.set_color("g") for i in plt.gca().get_xticklabels()]
[i.set_color("k") for i in plt.gca().get_yticklabels()]
ax.get_yaxis().set_visible(False)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(20) 
    tick.label.set_fontweight('bold')
#leg = plt.legend(framealpha = 0, loc = 'best')
#for text in leg.get_texts():
#    plt.setp(text, color = 'w',fontsize=24)
plt.savefig('gene2.pdf',dpi=600, bbox_inches='tight',transparent=True)
```

```python
plt.figure(figsize=(15,8))

sns.set_style('dark')
sns.set(rc={'axes.facecolor':'.4', 'figure.facecolor':'.3'})

ax=dfc.plot(kind='bar',color='y',legend=False)
ax.set_title('Cell Lines',color='w',fontsize=36,y=1.03)
[i.set_color("y") for i in plt.gca().get_xticklabels()]
[i.set_color("k") for i in plt.gca().get_yticklabels()]
ax.get_yaxis().set_visible(False)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(20) 
    tick.label.set_fontweight('bold')
#leg = plt.legend(framealpha = 0, loc = 'best')
#for text in leg.get_texts():
#    plt.setp(text, color = 'w',fontsize=24)
plt.savefig('cell.pdf',dpi=600, bbox_inches='tight',transparent=True)
```
